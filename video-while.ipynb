```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
import math
```

```python
def distancia (a,b,n=1): 
    sumatoria=np.sum(np.sqrt(a*b))
    medio= 1/np.sqrt(np.average(a) * np.average(b)*n*n) 
    return math.sqrt((1-(medio*sumatoria)))
```

```python
# video para leer
video= cv2.VideoCapture('360.mp4')
fps=video.get(cv2.CAP_PROP_FPS) 
cantFrames=video.get(cv2.CAP_PROP_POS_MSEC) #155 (hay 4 frames da√±ados)
size = (int(video.get(cv2.CAP_PROP_FRAME_WIDTH)), int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))) 

hbins=np.arange(0,179) # rango para canal H
vbins=np.arange(0,255) # rango para canal V

# Frame 1
i=0
success, frame = video.read()

hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
h1,s1,v1 = cv2.split(hsv); #h = hsv[:,:,0] 

#HIST1
hist1, bins = np.histogram(h1,bins=hbins,density=True)

while success:
    i+=1
    success, frame2 = video.read()
    if i <= 151:
        hsv2= cv2.cvtColor(frame2, cv2.COLOR_BGR2HSV)
        h2,s2,v2 = cv2.split(hsv2) 
        #HIST2
        hist2, bins = np.histogram(h2,bins=hbins,density=True)
        print(distancia(hist1,hist2, len(hbins)-1))
        hist1=hist2
    
```
